```python
# Read csv files:(comma separated files) it is a simple way to store big and very big data sets.csv files contains plain texts

# Loading the csv into a data frame with to_string()
import pandas as pd
df=pd.read_csv('cereal.csv')
# imp note always use single inverted commas for importing a file 
print(df.to_string())
```

```python
# Loading the csv into a data frame without to_string()
import pandas as pd
df=pd.read_csv('cereal.csv')
# imp note always use single inverted commas for importing a file 
print(df)
# if we dont use to_string() then the ide just shows us the topmost and bottommost 5 values and put .. on rest of the data
```

```python
# max_rows : you can check the maximum rows with :
import pandas as pd
print(pd.options.display.max_rows)
# the 60 output means any data frame or csv file which has no of rows more then 60 for that we have to view all the data through to_string() function
# print(pd.options.display.max_rows): this prints how many rows are supported in our computer
```

```python
# yes we can increase the maximum number of rows to display the entire dataframe 
import pandas as pd
pd.options.display.max_rows=9999
df=pd.read_csv('cereal.csv')
print(df)


# for best output always use to_string()
```
